### Modelling and predicting the clusters for the data

```python
#Scale the data first
scaler=StandardScaler()
df_scaled=scaler.fit_transform(df)
```

```python
#Check the shape of our scaled dataframe
df_scaled.shape
```

```python
#View our scaled dataframe
df_scaled
```

```python
#Finding the optimum number of clusters using the elbow method

scores_1=[]
range_values=range(1,20)

for i in range_values:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_scaled)
    scores_1.append(kmeans.inertia_)
    
plt.plot(scores_1)
```

```python
#Fitting our kmeans model
kmeans=KMeans(7)
kmeans.fit(df_scaled)
labels=kmeans.labels_ #label clusters associated to each data point

```

```python
#View the shape of our cluster_centers
kmeans.cluster_centers_.shape
```

```python
#Create a dataframe with our cluster centers
cluster_centers=pd.DataFrame(data=kmeans.cluster_centers_,columns=[df.columns])
```

```python
#View cluster_centers
cluster_centers
```

```python
# Reverse the scaling of data

cluster_centers=scaler.inverse_transform(cluster_centers)
cluster_centers=pd.DataFrame(data=cluster_centers,columns=[df.columns])

```

```python
#View cluster_centers with our scaling reversed.
cluster_centers
```

**Interpretation of clusters**:

A few distinct groups can be identfied below:

Customers with : 
* High purchase frequency, number of purchase transactions, amount, and high credit limit

* Low balance,low purchases amount ,low cash advance, percentage of full payment is relatively high at 24%

* High balance, high cash advance(>5k),high cash advance frequency and number of cash advance transactions, low percentage of full payment=3.8%

* low tenure of around 7 as compared to the other clusters with approximately 11+ years of tenure

```python
#Append the labels to our dataframe
df_cluster=df.copy()

df_cluster['Cluster']=labels

df_cluster
```

```python
#Plot the histograms for each clusters

for i in df.columns:
    plt.figure(figsize=(35,5))
    for j in range(7):
        plt.subplot(1,7,j+1)
        cluster=df_cluster[df_cluster['Cluster']==j]
        cluster[i].hist(bins=20)
        
        plt.title(f'{i}/ncluster{j}')


```

```python
#Fiting the PCA model 
pca=PCA(n_components=2)
principal_comp=pca.fit_transform(df_scaled)
principal_comp
```

```python
#Create a new dataframe using the pca model
pca_df=pd.DataFrame(data=principal_comp,columns=['PCA_1','PCA_2'])
```

```python
pca_df
```

```python
#Append the labels to our dataframe
pca_df_cluster=pca_df.copy()

pca_df_cluster['Cluster']=labels

pca_df_cluster
```
